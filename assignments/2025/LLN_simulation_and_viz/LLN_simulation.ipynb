```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
n = 10000000
plot_from_N = 50000

X = np.arange(1, n + 1)

tosses = np.random.randint(0, 2, n)  # 0 = tails, 1 = heads
averages = np.cumsum(tosses) / X

plt.plot(X[plot_from_N:], averages[plot_from_N:])
plt.axhline(0.5, color='red', linestyle='--')
plt.title("Law of Large Numbers in Action")
plt.xlabel("Number of Tosses")
plt.ylabel("Average of Heads")
plt.show()
```
