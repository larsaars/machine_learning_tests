```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x : x**2 + 1
df = lambda x : 2 * x

x = np.random.uniform(-5, 5)
eta = 0.00000005

history = [[x, f(x)]]

print(3 * '%12s ' % ('x', 'df(x)', 'f(x)'))

def log():
    print(f'{x:12.4f} {df(x):12.4f} {f(x):12.4f}') 
    
log()

for i in range(22):
    x -= eta * df(x)
    
    history.append([x, f(x)])
    # log() 
    
log()
    
points = np.array(history).reshape(2, -1)
```

```python
X = np.linspace(-5, 5, 1001)
Y, dY = f(X), df(X)

fig = plt.figure(figsize=(36, 21), dpi=72)
ax = plt.axes()

ax.plot(X, Y)
ax.plot(X, dY)

ax.scatter(*points, marker='o')

ax.grid(True, which='both')
 
plt.show()
```
